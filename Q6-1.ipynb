## Gradient Descent for finding the minimum of an arbitrary function

First of all, our arbitrary function $f(x)=7x^4-2x^3-91x^2-58x+24$

```python
func = lambda x: 7*(x**4)-2*(x**3)-91*(x**2)-58*(x)+24
```

```python
x = np.linspace(-4,4,100)
y = func(x)

plt.title("f(x)")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x,y);
```

For $f(x)$, we compute $f'(x) = 28x^3 - 6x^2 - 182x - 58$. 
Then we denote $x_0$ as the initial value, and set the learning rate $\eta$ = 0.01.  
The iteration of gradient descent is
$$x_{n+1} = x_n - \mbox{learning rate} \times \frac{dy}{dx} = x_n - \eta \times f'(x)$$

### Function of Gradient

```python
Gradient = lambda x: 28*(x**3)-6*(x**2)-182*x-58 
```

### Description of parameters in `Find_min`

- `x_init`: Where the algorithm starts
- `eta`: Learning rate
- `precision`: The acceptable difference between $x_{n+1}$ and $x_n$
- `diff`: Given initial difference 
- `cnt`: Count of iterations
- `max_cnt`: Max count of iterations 

```python
def Gradient_Find_min(x_init, eta, precision):
    diff = 1
    cnt = 0
    max_cnt = 100
    x_iter = []
    while diff > precision and cnt < max_cnt:
        x_prev = x_init
        x_init = x_init - eta * Gradient(x_prev) # Restart searching
        diff = abs(x_init - x_prev)
        cnt = cnt + 1
        x_iter.append(x_init)
    print("The minimum of the function occurs at", round(x_init,5))
    return np.array(x_iter)
```

```python
def Plot_Gradient_Descent(x_init, rate):
    x = np.linspace(-4,4,100)
    y = func(x)

    plt.title("f(x)")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.plot(x,y);

    x_iter = Gradient_Find_min(x_init, eta=rate, precision=0.00001)
    y_iter = func(x_iter)
    plt.scatter(x_iter, y_iter, c='r');
    
    #k = np.c_[x_iter, y_iter]
    #print("The iterations for (x,y)")
    #np.set_printoptions(suppress=True)
    #print(np.ndarray.round(k,5))
```
